## 1. Job Market

```python
#Continuing with the job market data, use Matplotlib/Seaborn to visualize:
```

```python
df = pd.read_csv('job-market.csv')
df.head()
```

```python
# Job by location
df1 = df.groupby(["Classification"])["Id"].count().reset_index(name="Number of Jobs")
df1.head()
```

```python
plt.figure(figsize=(15,8))

sns.set(style="darkgrid")

#ax = sns.barplot(x="budget2018", y="ministere", data=budget, label="Total")
sns.set_color_codes("pastel")
sns.barplot(x="Number of Jobs", y="Classification", data=df1)

plt.title('Number of Jobs by Classification')
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

```python
# Setting labels for items in Chart
Employee = ['Roshni', 'Shyam', 'Priyanshi',
            'Harshit', 'Anmol']
 
# Setting size in Chart based on
# given values
Salary = [40000, 50000, 70000, 54000, 44000]
 
# colors
colors = ['#FF0000', '#0000FF', '#FFFF00',
          '#ADFF2F', '#FFA500']
# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
 
# Pie Chart
plt.pie(Salary, colors=colors, labels=Employee,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
 
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
    
    

plt.figure(figsize=(15,8))

sns.set(style="darkgrid")

#ax = sns.barplot(x="budget2018", y="ministere", data=budget, label="Total")
sns.set_color_codes("pastel")
sns.barplot(x="Number of Jobs", y="Classification", data=df1)



plt.title('Job Posts by Salary Range')
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
