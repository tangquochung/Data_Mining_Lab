```python
import pandas as pd
import numpy as np
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Flight dataset
```

```python
df_flights = sns.load_dataset("flights")
```

```python
df_flights.head(5)
```

```python
may_flights = df_flights[df_flights['month']=='May']
sns.lineplot(data=may_flights, x='year', y="passengers")
```

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head(5)
```

```python
reduce_df_tips = df_tips[['tip','sex']]
```

```python
reduce_df_tips_mean = reduce_df_tips.groupby(['sex']).mean().reset_index()
reduce_df_tips_sum = reduce_df_tips.groupby(['sex']).sum().reset_index()
```

```python
f1 = plt.figure(1)
sns.barplot(data = reduce_df_tips_mean, x = "sex", y = "tip").set_title("mean tips")
f2 = plt.figure(2)
sns.barplot(data = reduce_df_tips_sum, x = "sex", y = "tip").set_title("total tips")
```

```python
reduce_df_tips_sum
```

```python
reduce_df_tips_sum.set_index('sex').plot.pie(y = 'tip', figsize = (5, 5), autopct = '%1.1f%%')
```
